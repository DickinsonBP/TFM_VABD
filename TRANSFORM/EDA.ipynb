# EDA de los datos extraidos de la Liga Femenina Endesa

## Librerias necesarias para el EDA

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from dotenv import load_dotenv
import os
```

## Cargar CSV en un dataframe

```python
df = pd.read_csv('LFENDESA_UnifiedData.csv',sep=';')
df.head(10)
```

```python
n_instances = df.shape[0]
input_features_info = df.dtypes.to_frame(name="Data Type").reset_index()
input_features_info.columns = ["Feature", "Data Type"]


print(f"Número de instancias totales: {n_instances}")
print(f"Atributos de entrada y su tipo:")
print(input_features_info)
print(f"\nValores nulos por cada variable:")
print(df.isnull().sum())
```

## Mostrar distribución de las variables

```python
# % T2
# % T3
# % T1
sns.set(style="whitegrid")

# Tamaño de los gráficos
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

sns.boxplot(y=df['% T2'],ax=axes[0], color='dodgerblue')
axes[0].set_title('Distribución de la Tiros libres')
axes[0].set_xlabel('')
axes[0].set_ylabel('T1')

# Reemplazado por boxplot
sns.boxplot(y=df['% T2'], ax=axes[1], color='dodgerblue')
axes[1].set_title('Distribución de la T2 Anotados')
axes[0].set_xlabel('')
axes[1].set_ylabel('T2')

# Reemplazado por boxplot
sns.boxplot(y=df['% T3'],ax=axes[2], color='dodgerblue')
axes[2].set_title('Distribución de la T3 Anotados')
axes[0].set_xlabel('')
axes[2].set_ylabel('T3')

# Ajustar los gráficos
plt.tight_layout()
plt.show()
```

```python
# Reb_Ofensivos             0
# Reb_Defensivos            0
# Rebotes                   0
sns.set(style="whitegrid")

# Tamaño de los gráficos
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

sns.boxplot(y=df['Reb_Ofensivos'],ax=axes[0], color='dodgerblue')
axes[0].set_title('Distribución de Rebotes Ofensivos')
axes[0].set_xlabel('')
axes[0].set_ylabel('Reb_Ofensivos')

# Reemplazado por boxplot
sns.boxplot(y=df['Reb_Defensivos'], ax=axes[1], color='dodgerblue')
axes[1].set_title('Distribución de Rebotes Defensivos')
axes[0].set_xlabel('')
axes[1].set_ylabel('Reb_Defensivos')

# Reemplazado por boxplot
sns.boxplot(y=df['Rebotes'],ax=axes[2], color='dodgerblue')
axes[2].set_title('Distribución de Rebotes')
axes[0].set_xlabel('')
axes[2].set_ylabel('Rebotes')

# Ajustar los gráficos
plt.tight_layout()
plt.show()
```

```python
# Asistencias               0
# Robos                     0
# Perdidas                  0
sns.set(style="whitegrid")

# Tamaño de los gráficos
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

sns.boxplot(y=df['Asistencias'], ax=axes[0], color='dodgerblue', width=0.3)
axes[0].set_title('Distribución de Asistencias')
axes[0].set_xlabel('')
axes[0].set_ylabel('Asistencias')

# Boxplot para Robos
sns.boxplot(y=df['Robos'], ax=axes[1], color='dodgerblue', width=0.3)
axes[1].set_title('Distribución de Robos')
axes[1].set_xlabel('')
axes[1].set_ylabel('Robos')

# Boxplot para Perdidas
sns.boxplot(y=df['Perdidas'], ax=axes[2], color='dodgerblue', width=0.3)
axes[2].set_title('Distribución de Pérdidas')
axes[2].set_xlabel('')
axes[2].set_ylabel('Pérdidas')

# Ajustar los gráficos
plt.tight_layout()
plt.show()
```

```python
# Tapones                   0
# Tapones Recibidos         0
sns.set(style="whitegrid")

# Tamaño de los gráficos
fig, axes = plt.subplots(1, 2, figsize=(15, 5))

sns.boxplot(y=df['Tapones'], ax=axes[0], color='dodgerblue', width=0.3)
axes[0].set_title('Distribución de Tapones')
axes[0].set_xlabel('')
axes[0].set_ylabel('Tapones')

# Boxplot para Tapones Recibidos
sns.boxplot(y=df['Tapones Recibidos'], ax=axes[1], color='dodgerblue', width=0.3)
axes[1].set_title('Distribución de Tapones Recibidos')
axes[1].set_xlabel('')
axes[1].set_ylabel('Tapones Recibidos')

# Ajustar los gráficos
plt.tight_layout()
plt.show()
```

```python
# Faltas Cometidas          0
# Faltas Recibidas          0
sns.set(style="whitegrid")

# Tamaño de los gráficos
fig, axes = plt.subplots(1, 2, figsize=(15, 5))

sns.boxplot(y=df['Faltas Cometidas'], ax=axes[0], color='dodgerblue', width=0.3)
axes[0].set_title('Distribución de Faltas Cometidas')
axes[0].set_xlabel('')
axes[0].set_ylabel('Faltas Cometidas')

# Boxplot para Faltas Recibidas
sns.boxplot(y=df['Faltas Recibidas'], ax=axes[1], color='dodgerblue', width=0.3)
axes[1].set_title('Distribución de Faltas Recibidas')
axes[1].set_xlabel('')
axes[1].set_ylabel('Faltas Recibidas')

# Ajustar los gráficos
plt.tight_layout()
plt.show()
```
